```python
from sklearn.datasets import load_iris
iris=load_iris()
x,y = iris.data,iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
from sklearn.svm import SVC
model = SVC(kernel='linear') #support vector classifier
model.fit(x_train,y_train) # train the model SVC with linear kernel
score= model.score(x_test,y_test)
print(f'model score: {score}')
y_pred= model.predict(x_test)
print(f'model prrediction:\n {y_pred}')
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix
acc= accuracy_score(y_test,y_pred)
print(f'accuracy score:\n {acc}')
mse=mean_squared_error(y_test,y_pred)
print(f'mean squared error:\n {mse}')
cm= confusion_matrix(y_test,y_pred)
print(f'Confusion matrix:\n {cm}')
import matplotlib.pyplot as plt
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred)
plt.plot(x_test[:,0],x_test[:,1])
plt.title('Scatter plot of x_test feature-0 v/s feature-1')
plt.xlabel('feature-0')
plt.ylabel('feature-1')
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
x,y = iris.data, iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
from sklearn.svm import SVC
model=SVC()
grid_params= {
    'C' : [0.1,1,10,100],
    'kernel' : ['linear','rbf','sigmoid','poly'],
    'gamma':['scale','auto', 0.001,0.01,0.1]
}
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(model,grid_params,cv=5)
grid.fit(x_train,y_train)
score= grid.score(x_test,y_test)
print(f'grid model score: {score}')
y_pred = grid.predict(x_test)
print(f'grid model prediction:\n {y_pred}')
print(f'grid best score:\n {grid.best_score_}')
print(f'grid best parametrs: {grid.best_params_}')
```

```python
from sklearn.datasets import load_iris
iris= load_iris()
x,y = iris.data, iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
from sklearn.svm import SVC
model= SVC()
import numpy as np 
random_params = {
    'C' : np.logspace(-2,2,20),
    'gamma': np.logspace(-3,1,20),
    'kernel': ['linear', 'rbf', 'poly','sigmoid']
}
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(model,random_params, cv=5, n_iter=10)
random.fit(x_train,y_train)
score= random.score(x_test,y_test)
print(f'random model score: {score}')
y_pred = random.predict(x_test)
print(f'random model prediction:\n {y_pred}')
print(f'random best score:\n {random.best_score_}')
print(f'random best parametrs: {random.best_params_}')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
x,y = iris.data,iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
from sklearn.svm import SVR
model = SVR(kernel='linear') #support vector regressor
model.fit(x_train,y_train) # train the model SVR with linear kernel
score= model.score(x_test,y_test)
print(f'model score: {score}')
y_pred= model.predict(x_test)
print(f'model prediction:\n {y_pred}')
from sklearn.metrics import r2_score, mean_squared_error 
r2= r2_score(y_test,y_pred)
print(f'r2 score:\n {r2}')
mse=mean_squared_error(y_test,y_pred)
print(f'mean squared error:\n {mse}')
import matplotlib.pyplot as plt
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred)
plt.plot(x_test[:,0],x_test[:,1])
plt.title('Scatter plot of x_test feature-0 v/s feature-1')
plt.xlabel('feature-0')
plt.ylabel('feature-1')
plt.show()
```
