```python
def display(name,dict):
    print(f'{name} of nodes of social network')
    for node,value in dict.items():
        print(f'node-{node}:{value}')
from networkx import Graph
g = Graph()
edges =[(1,2),(1,3),(2,3),(2,4),(3,4)]
g.add_edges_from(edges)
from networkx import draw
draw(g,node_size=500,node_color='green',with_labels=True)
import matplotlib.pyplot as plt
plt.title('Social Network Visualization')
plt.show()
from networkx import betweenness_centrality
display('betweenness centrality',betweenness_centrality(g))
from networkx import closeness_centrality
display('closeness centrality',closeness_centrality(g))
from networkx import degree_centrality
display('degree centrality',degree_centrality(g))
from networkx import pagerank
display('Pagerank',pagerank(g))

    
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
feature_vectors = iris.data
feature_names = iris.feature_names
species = iris.target
species_names = iris.target_names
import numpy as np
for i in range(4):
    feature = feature_vectors[:,i]
    print(f'\nSummary statistics of {feature_names[i]}')
    print(f'mean of {feature_names[i]}: {np.mean(feature)}')
    for j in range(3):
        print(f'mean of {feature_names[i]} of {species_names[j]}: {np.mean(feature[species==j])}')
    print(f'min of {feature_names[i]}: {np.min(feature)}')
    print(f'max of {feature_names[i]}: {np.max(feature)}')
    std = np.std(feature)
    print(f'Standard deviation of{feature_names[i]}: {std}')
    print(f'Variance of{feature_names[i]}: {std*std}')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
feature_names = iris.feature_names
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
labels = kmeans.fit_predict(x)
print(f'predicted labels:\n {labels}')
import matplotlib.pyplot as plt
for i in range(0,4,2):
    feature1 = x[:,i]
    feature2 = x[:,i+1]
    plt.scatter(feature1,feature2,c=labels)
    plt.plot(feature1,feature2)
    plt.title(f'Scatter plot of {feature_names[i]} v/s {feature_names[i+1]}')
    plt.xlabel(f'{feature_names[i]}')
    plt.ylabel(f'{feature_names[i+1]}')
    plt.show()
    
```
